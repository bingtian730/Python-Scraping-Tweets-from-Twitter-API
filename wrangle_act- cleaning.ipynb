```python
# open json file and show it as a dataframe 
import json
with open('tweet_json.txt')as json_file:
    data=json.load(json_file) 
df_tweet=pd.read_json(data)
df_tweet.head()
```

```python
df_tweet.info()
```

```python
df_tweet.isnull().sum()
```

```python
df_tweet.describe()
```

```python
# Bin edges that will be used to "cut" the data into groups
bin_favorite_edges=[0.0,5000.0,10000.0,20000.0,150000.0]
bin_favorite_names=['0-5000','5000-10000','10000-20000','20000-150000']
bin_retweet_edges=[0,2000,3500,6000,62000]
bin_retweet_names=['0-5000','5000-10000','10000-20000','20000-150000']
```

```python
# create columns that categorize retweet_count and favorite_count
df_tweet['favor_count'] = pd.cut(df_tweet['favorite_count'], bin_favorite_edges, labels=bin_favorite_names)
```

```python
df_tweet['retwe_count']=pd.cut(df_tweet['retweet_count'], bin_retweet_edges, labels=bin_retweet_names)
```

```python
df_tweet.head()
```

```python
# drop useless columns
df_tweet.drop(['favorited','full_text','retweeted'],axis=1, inplace=True)
```

```python
df_tweet.head()
```

```python
# notice there's null values in culumn
df_tweet.info()
```

```python
# rename column so that the data frames can merge
df_tweet.rename(index=str,columns={'id':'tweet_id'},inplace=True)
```

```python
# merge dataframes and create a new dataframe df_new
frames=[df_tweet,df_archive]
df_new=df_tweet.merge(df_archive)
```

```python
df_new.head()
```

```python
df_new.info()
```

```python
# find the null values
df_new.isnull().sum()
```

```python
# drop the null values
df_new.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
df_new.info()
```

```python
df_new.head()
```

```python
# find if there's duplicated rows
df_new.duplicated().sum()
```

```python
df_new['stage'] = 'doggo'
df_new.loc[df_new['floofer'] == 'floofer', 'stage'] = 'floofer'
df_new.loc[df_new['pupper'] =='pupper', 'stage'] = 'pupper'
df_new.loc[df_new['puppo']=='puppo','stage']='puppo'
```

```python
df_new.head()
```

```python
df_new.drop(['doggo','floofer','puppo','pupper'],axis=1, inplace=True)
```

```python
df_new.to_csv('new_clean.csv')
```
